#### Helper functions

```python
def get_dependency_types(names):
    spec_json = [json.load(open(os.path.join('apps', name, 'spec.json'), 'r')) for name in names] 
    # assuming only a single dependency per binder
    def populate_deps_dict(spec, deps_dict):
        deps = spec.get('dependencies')
        name = spec.get('name')
        dep = deps[0] if deps else None
        if dep in deps_dict:
            deps_dict[dep].append(name)
        else:
            deps_dict[dep] = [name]
    deps_dict = {}
    for spec in spec_json:
        populate_deps_dict(spec, deps_dict)
    return deps_dict
```

```python
def handle_requirements(app): 
    f = open(os.path.join('apps', app, 'repo', 'requirements.txt'), 'r')
    return map(lambda line: line.strip(), f.readlines())

def handle_environment(app): 
    f = open(os.path.join('apps', app, 'repo', 'environment.yml'), 'r')
    yml = yaml.load(f)
    def handle_pip(dep):
        return dep['pip']
    dep_list = [] 
    for dep in yml.get('dependencies', []):
        if isinstance(dep, dict) and 'pip' in dep:
            pip_deps = dep['pip']
            for dep in pip_deps:
                dep_list.append(dep.strip())
        else: 
            dep_list.append(dep.strip())
    return dep_list 

def get_dependencies(names):
    dep_types = get_dependency_types(names)
    deps = {}
    
    # get the unprocessed dependencies for each dependency type
    for dep, apps in dep_types.items():
        if dep == 'requirements.txt':
            deps[dep] = map(lambda app: handle_requirements(app), apps)
        elif dep == 'environment.yml':
            deps[dep] = map(lambda app: handle_environment(app), apps)
        elif dep == 'Dockerfile':
            # won't look into the Dockerfiles for now
            pass
    
    # flatten the lists
    dep_dict = dict(map(lambda (name, l): (name, [i for sublist in l for i in sublist]), deps.items()))
    
    # remove invalid lines and version numbers
    git_re = re.compile('git\+https\:\/\/github.com\/.*/(?P<name>.*).git')
    def is_valid(line):
        stripped = line.strip()
        return not stripped.startswith('#') and stripped != '\\n'
    def just_name(line):
        if line.startswith('git+'):
            return git_re.search(line).group('name')
        name = line.split('=')[0]
        for char in ['=', '>', '<']: 
            if char in name:
                name = name.replace(char, '')
        return name
    for dep in dep_dict:
        lines = dep_dict[dep]
        dep_dict[dep] = map(just_name, filter(is_valid, lines)) 
    return dep_dict
```
